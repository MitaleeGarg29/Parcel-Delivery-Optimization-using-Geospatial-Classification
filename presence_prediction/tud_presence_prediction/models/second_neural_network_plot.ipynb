```python
import pandas as pd
import matplotlib.pyplot as plt
import wandb

wandb.login()
```

Load the data metrics file

```python
# Load your metrics data
file_path = "D:/TU Darmstadt/Thesis/presence_prediction/tud_presence_prediction/training_results/second_neural_network/version_0/metrics_all.csv"
metrics_df = pd.read_csv(file_path)

# Print available columns to check the correct column name
print("Available columns:", metrics_df.columns)
```

# Plot Loss for each users
for Day 0, 1 and 2 for both train mode

```python
# Get unique users
users = metrics_df["user"].unique()

# Create a figure for subplots (1 row: Train Mode, 3 columns: Day 0, Day 1, Day 2)
fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharey=True)

# Days to plot
days = [0, 1, 2]

# Iterate over days and create subplots
for col, day in enumerate(days):
    day_data = metrics_df[metrics_df["day"] == day]

    # Plot Train Mode
    for user in users:
        user_data = day_data[(day_data["user"] == user) & (day_data["mode"] == "train")]
        axes[col].plot(user_data["epoch"], user_data["loss"], label=f"User {user}")
    axes[col].set_title(f"Loss vs. Epochs (Train Mode) - Day {day}")
    axes[col].set_xlabel("Epochs")
    if col == 0:
        axes[col].set_ylabel("Loss")
    axes[col].grid(True)
    axes[col].legend(loc="upper right")  # Add legends for all subplots

# Adjust layout
plt.tight_layout()
plt.show()
```

# Optimal Threshold 
For Day 0 subploting 13 users

```python
import matplotlib.pyplot as plt

# Filter the data for Day 0 and Train Mode
day_0_data = metrics_df[(metrics_df["day"] == 0) & (metrics_df["mode"] == "train")]

# Get unique users
users = day_0_data["user"].unique()

# Create subplots for 13 users
fig, axes = plt.subplots(4, 4, figsize=(16, 16))  # 4x4 grid (adjust based on the number of users)
axes = axes.flatten()  # Flatten axes for easier indexing

# Iterate over users and create a plot for each
for i, user in enumerate(users):
    user_data = day_0_data[day_0_data["user"] == user]
    ax = axes[i]
    ax.plot(user_data["epoch"], user_data["f1_score"], label=f"User {user}")
    ax.set_title(f"User {user} - Day 0")
    ax.set_xlabel("Epochs")
    ax.set_ylabel("F1 Score")
    ax.grid(True)
    ax.legend(loc="lower right")

# Hide unused subplots if fewer than 16 users
for j in range(len(users), len(axes)):
    axes[j].set_visible(False)

# Adjust layout
plt.tight_layout()
plt.show()
```
