```python
import numpy as np
import matplotlib.pyplot as plt

# Define true label (y) and predicted probability (y_hat)
y = np.array([1, 0])  # True labels: 1 for "home", 0 for "not home"
y_hat = np.linspace(0.01, 0.99, 100)  # Predicted probabilities

# Compute BCE loss for true labels
bce_loss_y1 = -np.log(y_hat)  # When y = 1
bce_loss_y0 = -np.log(1 - y_hat)  # When y = 0

# Weighted BCE (example weights: w1 = 2 for y=1, w0 = 1 for y=0)
w1 = 2  # Weight for positive class
w0 = 1  # Weight for negative class
weighted_bce_loss_y1 = -w1 * np.log(y_hat)  # Weighted BCE for y=1
weighted_bce_loss_y0 = -w0 * np.log(1 - y_hat)  # Weighted BCE for y=0

# Plot BCE loss curves
plt.figure(figsize=(12, 6))

# Plot regular BCE loss
plt.subplot(1, 2, 1)
plt.plot(y_hat, bce_loss_y1, label='BCE Loss (y=1)', color='blue')
plt.plot(y_hat, bce_loss_y0, label='BCE Loss (y=0)', color='orange')
plt.title('BCE Loss for Correct and Incorrect Predictions')
plt.xlabel('Predicted Probability (y_hat)')
plt.ylabel('BCE Loss')
plt.legend()
plt.grid(True)

# Plot weighted BCE loss
plt.subplot(1, 2, 2)
plt.plot(y_hat, weighted_bce_loss_y1, label='Weighted BCE Loss (y=1)', color='blue')
plt.plot(y_hat, weighted_bce_loss_y0, label='Weighted BCE Loss (y=0)', color='orange')
plt.title('Weighted BCE Loss for Imbalanced Classes')
plt.xlabel('Predicted Probability (y_hat)')
plt.ylabel('Weighted BCE Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
